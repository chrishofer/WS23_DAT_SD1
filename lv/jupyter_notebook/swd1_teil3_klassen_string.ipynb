Ein *Modul (module)* ist eine gruppierte Sammlung von Variablen und Funktionen in einer Datei. blicherweise beziehen sich die Inhalte aufeinander. Ein Modul kann mittels *import* importiert werden.
```python

```

````python
import math

print(type(math)) # gibt wirklich eine neue var vom typ module
#help(math)


# etwas aus dem modul verwenden - zugriff mit dem namen
print(math.pi)
print(math.sqrt(25))

print(math.ceil(7.2)) # ceil funktion aus dem mathe modul

def ceil(para):
    print("bla bla ceil bla")

    
ceil(7.2)
````

Ein Objekt (object) oder auch Instanz (instance) genannt ist eine konkrete Realisierung einer Klasse. Ein Glas Apfelsaft beispielsweise besteht nur aus einem Bestandteil Apfelsaft, hat eine gewisse Menge bzw. einen Preis und man kann damit konkrete Aktionen (Methoden) ausfhren - zum Beispie zubereiten oder trinken. Ein Glas Wasser hat als zweites Objekt hat genauso eigene Eigenschaften und ausgefhrte Aktionen betreffen nur genau dieses Objekt. 




Zurck zu Python: Jeder Typ in Python ist eigentlich eine Klasse (class). Deshalb haben wir jetzt schon immer von Objekten geredet ohne eigentlich darber zu sprechen was genau gemeint war. Funktionen die sich auf Objekte einer Klasse beziehen werden *Methoden* genannt. Der Unterschied ist, dass der erste Parameter einer Methode eine konkrete Instanz der Klasse ist.
#Schauen wir uns einmal den Typ (die Klasse) String in der Hilfe an:

````python
#help(str)


name = "hansi"

# das ist schon komfortvariante 
print(name.capitalize())

# was passiert eigentlich (machen wir aber nicht - passiert aber im hintergrund)
print(str.capitalize(name)) # von der klasse str wird die methode capitalize aufgerufen - erster parameter ist objekt/instanz der klasse str und wird mit self "angesprochen"

#  center(self, width, fillchar=' ', /)

print(name.center(10))
print(name.center(20, "*"))
````

````python
# UE 1) Erstellen Sie eine Funktion taschenrechner, die einen String 
# als Parameter erhlt. Mit folgenden Eingaben soll der Taschenrechner
# umgehen knnen, wobei <zahl> bzw. <zahl2> fr eine beliebige ganzzahlige 
# Zahl steht: 
# "zahl + zahl2"  bzw. alternativ  "zahl plus zahl2"
# "zahl - zahl2"  bzw. alternativ  "zahl minus zahl2"

# "zahl * zahl2"  bzw. alternativ  "zahl mal zahl2"
# "zahl / zahl2"  bzw. alternativ  "zahl durch zahl2" 

# Die funktion soll das Ergebnis der Operation als Zahl zurckliefern bzw. None
# falls kein passender Operator angegeben wurde.
# Hinweis: Das ist eine eher simple Lsung - es gibt elegantere und krzere Lsungsmglichkeiten
# und es werden nicht alle Fehlerflle abgefangen


#x = taschenrechner("18 + 12")
#x = taschenrechner("18 plus 12")
#x = taschenrechner("18 durch 12")
# auf x soll dann die zahl 30 sein
                   
test = "12 plus 13"
test2 = "12, 13, 14, 15"
# beides richtig
print(test.partition("plus")) # teilt in tupel auf (3 teilung)
print(test2.split(",")) # split nimmt zeichen und teilt beliebig viele inhalte auf und gibt sie in liste
print(test2.partition(",")) # partition teilt immer in 3 teile auf

# Lsung

def taschenrechner(rechnung : str) -> int:   
    # +
    pos = rechnung.find("+")
    if pos != -1:
        return int(rechnung[:pos]) + int(rechnung[pos + len("+"):])
    pos = rechnung.find("plus")
    if pos != -1:
        return int(rechnung[:pos]) + int(rechnung[pos + len("plus"):])
    # -
    pos = rechnung.find("-")
    if pos != -1:
        return int(rechnung[:pos]) - int(rechnung[pos + len("-"):])
    pos = rechnung.find("minus")
    if pos != -1:
        return int(rechnung[:pos]) - int(rechnung[pos + len("minus"):])
    # *
    pos = rechnung.find("*")
    if pos != -1:
        return int(rechnung[:pos]) * int(rechnung[pos + len("*"):])
    pos = rechnung.find("mal")
    if pos != -1:
        return int(rechnung[:pos]) * int(rechnung[pos + len("mal"):])
    # /
    pos = rechnung.find("/")
    if pos != -1:
        return int(rechnung[:pos]) / int(rechnung[pos + len("/"):])
    pos = rechnung.find("durch")
    if pos != -1:
        return int(rechnung[:pos]) / int(rechnung[pos + len("durch"):])
    
    return None; 

print(taschenrechner("  12 + 13  "))
print(taschenrechner("  12 plus 13  "))
print(taschenrechner("  12 - 13  "))
print(taschenrechner("  12 minus 13  "))
print(taschenrechner("  12 * 2  "))
print(taschenrechner("  12 mal 2  "))
print(taschenrechner("  12 / 3  "))
print(taschenrechner("  12 durch 3  "))
print(taschenrechner("  12 test 13  "))

````
