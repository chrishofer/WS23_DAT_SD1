Ein *Modul (module)* ist eine gruppierte Sammlung von Variablen und Funktionen in einer Datei. Üblicherweise beziehen sich die Inhalte aufeinander. Ein Modul kann mittels *import* importiert werden.
```python
import math
```
Damit wird eine Variable mit dem Namen des Moduls erzeugt. (Diese ist vom Typ module)

````python
import math

print(type(math)) # gibt wirklich eine neue var vom typ module
#help(math)


# etwas aus dem modul verwenden - zugriff mit dem namen
print(math.pi)
print(math.sqrt(25))

print(math.ceil(7.2)) # ceil funktion aus dem mathe modul

def ceil(para):
    print("bla bla ceil bla")

    
ceil(7.2)
````

Mit dem Operator . greifen wir auf das Objekt links davon zu und suchen innerhalb dieses Objekts nach dem Namen rechts vom Punkt. Also konkret zuvor auf das Modulobjekt *math* und rufen von diesem Objekt die Methode *sqrt* auf.


Es gibt auch einen bequemeren Weg falls wir nur ausgewählte Teile aus einem Modul benötigen. In diesem Fall wird keine Modulvariable erzeugt sondern nur Variablen für die spezifizierten Variablen und Funktionen. (Achung falls schon eine solche Variable/Funktion mit diesem Namen im Namespace besteht wird sie überschrieben.)

### Klassen

Klassen können als Baupläne oder Schablonen gesehen werden. Diese beschreiben Eigenschaften als auch mögliche Aktionen (Methoden genannt). 


Falls wir beispielsweise eine Klasse *Getränk* für eine Restaurantsoftware schreiben:

Was wären mögliche Eigenschaften?

Was wären mögliche Methoden (Aktionen)?





Ein Objekt (object) oder auch Instanz (instance) genannt ist eine konkrete Realisierung einer Klasse. Ein Glas Apfelsaft beispielsweise besteht nur aus einem Bestandteil Apfelsaft, hat eine gewisse Menge bzw. einen Preis und man kann damit konkrete Aktionen (Methoden) ausführen - zum Beispie zubereiten oder trinken. Ein Glas Wasser hat als zweites Objekt hat genauso eigene Eigenschaften und ausgeführte Aktionen betreffen nur genau dieses Objekt. 




Zurück zu Python: Jeder Typ in Python ist eigentlich eine Klasse (class). Deshalb haben wir jetzt schon immer von Objekten geredet ohne eigentlich darüber zu sprechen was genau gemeint war. Funktionen die sich auf Objekte einer Klasse beziehen werden *Methoden* genannt. Der Unterschied ist, dass der erste Parameter einer Methode eine konkrete Instanz der Klasse ist.
#Schauen wir uns einmal den Typ (die Klasse) String in der Hilfe an:

````python
#help(str)


name = "hansi"

# das ist schon komfortvariante 
print(name.capitalize())

# was passiert eigentlich (machen wir aber nicht - passiert aber im hintergrund)
print(str.capitalize(name)) # von der klasse str wird die methode capitalize aufgerufen - erster parameter ist objekt/instanz der klasse str und wird mit self "angesprochen"

#  center(self, width, fillchar=' ', /)

print(name.center(10))
print(name.center(20, "*"))
````

Da jede Methode einer Klasse als ersten Parameter ein konkretes Objekt dieser Klasse benötigt, gibt es eine Kurzschreibweise mittels Punkt:

````python
# UE 1) Erstellen Sie eine Funktion taschenrechner, die einen String 
# als Parameter erhält. Mit folgenden Eingaben soll der Taschenrechner
# umgehen können, wobei <zahl> bzw. <zahl2> für eine beliebige ganzzahlige 
# Zahl steht: 
# "zahl + zahl2"  bzw. alternativ  "zahl plus zahl2"
# "zahl - zahl2"  bzw. alternativ  "zahl minus zahl2"

# "zahl * zahl2"  bzw. alternativ  "zahl mal zahl2"
# "zahl / zahl2"  bzw. alternativ  "zahl durch zahl2" 

# Die funktion soll das Ergebnis der Operation als Zahl zurückliefern bzw. None
# falls kein passender Operator angegeben wurde.
# Hinweis: Das ist eine eher simple Lösung - es gibt elegantere und kürzere Lösungsmöglichkeiten
# und es werden nicht alle Fehlerfälle abgefangen


#x = taschenrechner("18 + 12")
#x = taschenrechner("18 plus 12")
#x = taschenrechner("18 durch 12")
# auf x soll dann die zahl 30 sein
                   
test = "12 plus 13"
test2 = "12, 13, 14, 15"
# beides richtig
print(test.partition("plus")) # teilt in tupel auf (3 teilung)
print(test2.split(",")) # split nimmt zeichen und teilt beliebig viele inhalte auf und gibt sie in liste
print(test2.partition(",")) # partition teilt immer in 3 teile auf

# Lösung

def taschenrechner(rechnung : str) -> int:   
    # +
    pos = rechnung.find("+")
    if pos != -1:
        return int(rechnung[:pos]) + int(rechnung[pos + len("+"):])
    pos = rechnung.find("plus")
    if pos != -1:
        return int(rechnung[:pos]) + int(rechnung[pos + len("plus"):])
    # -
    pos = rechnung.find("-")
    if pos != -1:
        return int(rechnung[:pos]) - int(rechnung[pos + len("-"):])
    pos = rechnung.find("minus")
    if pos != -1:
        return int(rechnung[:pos]) - int(rechnung[pos + len("minus"):])
    # *
    pos = rechnung.find("*")
    if pos != -1:
        return int(rechnung[:pos]) * int(rechnung[pos + len("*"):])
    pos = rechnung.find("mal")
    if pos != -1:
        return int(rechnung[:pos]) * int(rechnung[pos + len("mal"):])
    # /
    pos = rechnung.find("/")
    if pos != -1:
        return int(rechnung[:pos]) / int(rechnung[pos + len("/"):])
    pos = rechnung.find("durch")
    if pos != -1:
        return int(rechnung[:pos]) / int(rechnung[pos + len("durch"):])
    
    return None; 

print(taschenrechner("  12 + 13  "))
print(taschenrechner("  12 plus 13  "))
print(taschenrechner("  12 - 13  "))
print(taschenrechner("  12 minus 13  "))
print(taschenrechner("  12 * 2  "))
print(taschenrechner("  12 mal 2  "))
print(taschenrechner("  12 / 3  "))
print(taschenrechner("  12 durch 3  "))
print(taschenrechner("  12 test 13  "))

````
